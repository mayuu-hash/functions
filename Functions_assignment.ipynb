# 6. Explain the concept of generators in Python and how they are defined.
Ans - Generators are special functions that return values one at a time using the yield keyword. They are used for creating iterators in a memory-efficient way.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

```python
def gen_numbers(n):
    for i in range(n):
        yield i
```

# 10. What is the difference between map(), reduce(), and filter() functions in Python?
Ans - map() applies function to all items. filter() selects items that meet condition. reduce() applies a rolling computation to the iterable( from functools).


```python
from functools import reduce

# map
list(map(lambda x: x + 1, [1, 2]))

# filter
list(filter(lambda x: x % 2 == 0, [1,2,3]))

# reduce
reduce(lambda x, y: x + y, [1,2,3])
```

# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

nums = [1,2,3,4,5,6]
result = sum_even_numbers(nums)
print("Sum of even numbers:", result)
```

# 4. Write a Python function that checks if a given number is prime or not from 1 to 100.

```python
def mayu(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(mayu(7))
print(mayu(10))
```

# 5. Create a iterator class in Python that generates the fibonacci sequence up to a specific number of terms.

```python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

fib = Fibonacci(10)

for num in fib:
    print(num)
```

# 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line

# Create a dummy file for testing
with open("sample.txt", "w", encoding="utf-8") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")


# Example usage to print the lines (test)
for line in read_file_lines("sample.txt"):
    print(line.strip())
```

# 10. Create a Python program that uses filter() to remove all vowels from a given string.

```python
def mayu(s):
    return ''.join(filter(lambda c: c.lower() not in "aeiou", s))

print(mayu("King In The North"))
```
